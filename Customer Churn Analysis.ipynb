```python
#Create a visualisation (Bar chart) to show how many customers joined before 2021 and after 2021.

# Convert JoinDate to datetime
Customers_df['JoinedDate'] = pd.to_datetime(Customers_df['JoinDate'], format="%d-%m-%Y")

# Define the date for comparison
date_threshold = pd.Timestamp("2021-01-01")

# Create a new column to categorize the join dates
Customers_df['JoinPeriod'] = Customers_df['JoinedDate'].apply(
    lambda x: "After" if x >= date_threshold else "Before"
)

# Display first 10 rows
print(Customers_df.head(10))




```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='JoinPeriod',data=Customers_df)
plt.title("Numbers of customers by join period: After / Before 2021")
plt.xlabel("Join Period")
plt.ylabel("Number of Customers")
plt.show()
print(Customers_df['JoinPeriod'].value_counts())
```

```python


# Assuming you have a column 'PlanType' that indicates whether it's 'Annual' or 'Other'
plan_counts = Subscription_df['PlanType'].apply(
    lambda x: 'Annual' if x == 'Annual' else 'Non-Annual'
).value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(plan_counts, labels=plan_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Subscriptions: Annual vs Non-Annual")
plt.show()

# Bar chart
plt.figure(figsize=(6,4))
plan_counts.plot(kind='bar', color=['skyblue','lightgreen'])
plt.title("Subscriptions: Annual vs Non-Annual")
plt.xlabel("Plan Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python


# Categorize transaction amounts
Transactions_df['AmountCategory'] = Transactions_df['Amount'].apply(
    lambda x: "Below 100" if x <= 100 else "Above 100"
)

# Count categories
Amount_counts = Transactions_df['AmountCategory'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(Amount_counts, labels=Amount_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Transactions: Above 100 vs Below 100")
plt.show()


#bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
Amount_counts.plot(kind='bar', color=['skyblue','lightgreen'])
plt.title("Transactions: Above 100 and Below 100")  
plt.xlabel("Amount Category")
plt.ylabel("Count")
plt.show()



    
```

```python
churn_counts = Churn_df['ChurnID'].value_counts()
```

```python
total_Customers = Customers_df.shape[0]
total_Churned_Customers = Churn_df.shape[0]
total_non_Customers = total_Customers-total_Churned_Customers
#create a dataframe to represent the two categories
Churn_status = pd.DataFrame({"Status": ["churned","Not Churned"],"Count":[total_Churned_Customers,total_non_Customers]})
```

```python
print(Churn_status)
```

```python
plt.figure(figsize=(8,6))
ax = sns.barplot(x="Status",y="Count",data=Churn_status)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churned vs Not Churned")
plt.xlabel("Customers Status")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.grid(True)
plt.show()
```
